```python
## importing modules
import pandas as pd 
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx', sheet_name=1, index_col='ID')
df.head(10)
```

```python
df.columns
```

```python
df = df[['Age', 'Experience', 'Income', 'ZIP Code', 'Family', 'CCAvg',
       'Education', 'Mortgage', 'Securities Account',
       'CD Account', 'Online', 'CreditCard', 'Personal Loan']]
df
```

```python
x = df.iloc[:,:-1]
```

```python
x
```

# Variable splitting:

```python
y = df.iloc[:,-1:]
```

```python
y['Personal Loan'].value_counts()
```

```python
## adding constant
x1 = sm.add_constant(x)
```

```python
logistic = sm.Logit(y,x1)
```

```python
result = logistic.fit()
```

```python
print(result.summary())
```

# Observation from the Logit table:

1. Most significant variables (P-value =0): Income,Family,Education,CD Account,Online,Credit Card
2. Significant varibles (P-value < 0.05) : CCAvg, Securities Account
3. nost so significant variables (P-value > 0.05): Age,Experience,Zip code,Mortgage. 

```python
## train_test_split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size = .3, random_state = 1)
```

```python
## logistic regression model
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_predict = log.predict(x_test)
```

```python
print('Accuracy:', log.score(x_test,y_test))
```

# conclusion:

Our model is 94.4 % accurate.

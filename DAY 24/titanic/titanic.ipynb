```python
## importing modules
import pandas as pd 
import numpy as np
from sklearn import preprocessing
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
titanic_data = pd.read_csv('train.csv')
titanic_data.head()
```

```python
## changing sex,embarked to int categoty

l = preprocessing.LabelEncoder()
titanic_data['Sex'] = l.fit_transform(titanic_data['Sex'])
titanic_data['Embarked'] = l.fit_transform(titanic_data['Embarked'])
titanic_data
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp',
        'Fare', 'Embarked']
```

```python
## random forest model
rf_model = RandomForestClassifier(n_estimators=1000,oob_score=True,max_features=2)
rf_model.fit(X= titanic_data[features],y= titanic_data['Survived'])
```

```python
for features,imp in zip(features,rf_model.feature_importances_):
    print(features,imp);
```

# Inference:
Out of the above features the most important are: 'Sex','Age','Fare'!!

```python
## decision tree based on these three features:

ID_VAR = pd.DataFrame([titanic_data['Sex'],titanic_data['Age'],titanic_data['Fare']]).T
tree_model = tree.DecisionTreeClassifier(max_depth=6)
tree_model.fit(X = ID_VAR,y= titanic_data['Survived'])
```

```python
## exporting decision tree:
with open ('Dtree.dot','w') as f:
    f= tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f);
    
```

```python
tree_model.score(X= ID_VAR,y= titanic_data['Survived'])
```

# Testing:

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
test_data['Sex'] = l.fit_transform(test_data['Sex'])
test_data['Embarked'] = l.fit_transform(test_data['Embarked'])
test_data.head()
```

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.naive_bayes import *
```

```python
train = pd.read_csv('train.csv',index_col='PassengerId')
train.head()
```

```python
l = preprocessing.LabelEncoder()
for i in ['Sex','Embarked']:
    train[i] = l.fit_transform(train[i])
```

```python
train = train.applymap(int)
train.info()
```

```python
features = ['Survived','Fare','Age','Pclass','Sex','SibSp','Parch','Embarked']
score_card = pd.DataFrame(data=[], columns=['features','Accuracy'])

for i in features:
    y = train[i]
    x = train.drop([i], axis = 1)
    x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = .3,random_state = 0)
    clf = BernoulliNB()
    y_predict = clf.fit(x_train,y_train).predict(x_test)
    acc_score = accuracy_score(y_test,y_predict)
    conf_mat = confusion_matrix(y_test,y_predict)
    score_card.loc[len(score_card)] = [i, acc_score]
    print('The confusion matrix for', i ,'is:', conf_mat)
```

```python
score_card.sort_values(['Accuracy'], ascending= False)
```
